```python
import sys
import Metashape
import os
import re
import json
from pathlib import Path
import pandas as pd
```

```python
# input file path
path0 = 'G:/island'

trips = ['2014_04', '2014_10', '2015_05', '2015_11', '2016_11', '2017_11', '2018_11', '2019_11', '2021_03', '2022_04', '2022_11', '2023_11']
sites = ['cooks_path', 'corner_beach', 'easter_point', 'gnarly_tree', 'horseshoe', 'lagoon_1', 'lagoon_2', 'lizard_head', 'no_mans_land', 'north_of_paradise', 'north_reef_1', 'north_reef_2', 'north_reef_3', 'osprey', 'resort', 'south_island', 'southeast', 'trimodal', 'turtle_beach', 'vickis', 'washing_machine']

# Create DataFrame to store results
df = pd.DataFrame(columns=['trip', 'site', 'images', 'images_eng', 'metashape_files', 'products', 'products_old', 'dem_res'])

# Function to check for specific file extensions in a directory
def check_for_files(directory, extensions):
    if os.path.exists(directory):
        for fname in os.listdir(directory):
            if fname.endswith(extensions):
                return True
    return False

# Loop through each trip and site to check folders
for t in trips:
    for s in sites:
        images_dir = os.path.join(path0, t, s, 'images')
        images_eng_dir = os.path.join(path0, t, s, 'images_eng')
        metashape_files_dir = os.path.join(path0, t, s, 'metashape_files')
        products_dir = os.path.join(path0, t, s, 'products')
        products_old_dir = os.path.join(path0, t, s, 'products_old')

        # Check if the required files exist in each directory
        images = check_for_files(images_dir, ('.jpg', '.png'))
        images_eng = check_for_files(images_eng_dir, ('.jpg', '.png'))
        metashape_files = "no file"
        
        # Check for metashape files and folder content
        if os.path.exists(metashape_files_dir):
            psx_files = [f for f in os.listdir(metashape_files_dir) if f.endswith('.psx') and not f.startswith('.')]
            if psx_files:
                # 
                total_files = len([f for f in os.listdir(metashape_files_dir) if not f.startswith('.')])
                if total_files > 2:
                    metashape_files = "ERROR"
                else:
                    metashape_files = "exists"
        
        products = check_for_files(products_dir, '.tif')
        products_old = check_for_files(products_old_dir, '.tif')

        # Assign a default None value to dem_res (since it's not calculated here)
        res = None

        # Append the results to the DataFrame
        df = pd.concat([df, pd.DataFrame({
            'trip': [t],
            'site': [s],
            'images': [images],
            'images_eng': [images_eng],
            'metashape_files': [metashape_files],
            'products': [products],
            'products_old': [products_old],
            'dem_res': [res]
        })], ignore_index=True)

# Save the DataFrame to a CSV file
df.to_csv('output.csv', index=False)
```

```python
# Filter for metashape_files == ERROR
metashape_false = df[(df['metashape_files'] == "ERROR")][['trip', 'site']]
print("Metashape files error:")
print(metashape_false)
```

```python
# Create DataFrame to store results
df = pd.DataFrame(columns=['trip', 'site', 'orthomosaic_resolution', 'DEM_resolution', 'dense_cloud_quality', 'aligned_photos'])

# Function to extract Metashape project data
def extract_metashape_data(doc, t, s):
    chunk = doc.chunk

    orthomosaic_resolution = None
    DEM_resolution = None
    dense_cloud_quality = None
    aligned_photos = None

    if chunk.orthomosaic:
        orthomosaic_resolution = chunk.orthomosaic.resolution  # e.g., in mm/pixel

    if chunk.elevation:
        DEM_resolution = chunk.elevation.resolution  # e.g., in mm/pixel

    # Check for dense cloud quality
    if hasattr(chunk, 'dense_cloud') and chunk.dense_cloud:
        if 'dense_cloud/depth_quality' in chunk.dense_cloud.meta:
            quality_mapping = {
                'Ultra': 'Ultra High',
                'High': 'High',
                'Medium': 'Medium',
                'Low': 'Low',
                'Lowest': 'Lowest'
            }
            dense_cloud_quality = quality_mapping.get(chunk.dense_cloud.meta['dense_cloud/depth_quality'], 'Unknown')
        else:
            dense_cloud_quality = 'Unknown'

    if chunk.cameras:
        aligned_photos = sum(1 for camera in chunk.cameras if camera.transform is not None)  # Number of aligned photos

    return {
        'trip': t,
        'site': s,
        'orthomosaic_resolution': orthomosaic_resolution,
        'DEM_resolution': DEM_resolution,
        'dense_cloud_quality': dense_cloud_quality,
        'aligned_photos': aligned_photos
    }

# Loop through each trip and site to check folders
for t in trips:
    for s in sites:
        metashape_files_dir = os.path.join(path0, t, s, 'metashape_files')
        
        if os.path.exists(metashape_files_dir):
            metashape_files_found = False
            for fname in os.listdir(metashape_files_dir):
                if fname.endswith('.psx') and not fname.startswith('.'):
                    metashape_files_found = True
                    try:
                        doc = Metashape.Document()
                        doc.open(os.path.join(metashape_files_dir, fname))
                        data = extract_metashape_data(doc, t, s)
                        df = df.append(data, ignore_index=True)
                    except Exception as e:
                        #  "error"
                        df = df.append({
                            'trip': t,
                            'site': s,
                            'orthomosaic_resolution': 'error',
                            'DEM_resolution': 'error',
                            'dense_cloud_quality': 'error',
                            'aligned_photos': 'error'
                        }, ignore_index=True)
            
            if not metashape_files_found:
                #  .psx  "no file"
                df = df.append({
                    'trip': t,
                    'site': s,
                    'orthomosaic_resolution': 'no file',
                    'DEM_resolution': 'no file',
                    'dense_cloud_quality': 'no file',
                    'aligned_photos': 'no file'
                }, ignore_index=True)
        else:
            #  "no file"
            df = df.append({
                'trip': t,
                'site': s,
                'orthomosaic_resolution': 'no file',
                'DEM_resolution': 'no file',
                'dense_cloud_quality': 'no file',
                'aligned_photos': 'no file'
            }, ignore_index=True)

# Save the DataFrame to a CSV file
df.to_csv('metashape_analysis_output.csv', index=False)
```
